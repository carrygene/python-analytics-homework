```python
#Перед початком
import pandas as pd

# Якщо файли в тій самій теці, що і ноутбук
df_public = pd.read_csv('survey_results_public.csv', low_memory=False)
df_schema = pd.read_csv('survey_results_schema.csv')

print(f"Основні дані: {df_public.shape}")
print(f"Опис структури: {df_schema.shape}")
print("\nПерші 5 рядків:")
print(df_public.head())
```

```python
#Завдання 1
# Загальна кількість респондентів
total_respondents = len(df_public)
print(total_respondents)
```

```python
#Завдання 2
# список питань з schema
questions = set(df_schema['qname'])

# список колонок з основних даних
columns = set(df_public.columns)

# спільні колонки (питання, на які були відповіді)
common_columns = questions.intersection(columns)

# підрахунок респондентів без пропусків у цих колонках
complete_responses = df_public[list(common_columns)].dropna()
complete_respondents = len(complete_responses)

print(complete_respondents)
```

```python
#Завдання 3

# Статистика для досвіду роботи (WorkExp)
mean_exp = df_public['WorkExp'].mean()
median_exp = df_public['WorkExp'].median()
mode_exp = df_public['WorkExp'].mode().iloc[0]

print("Середнє:", mean_exp)
print("Медіана:", median_exp)
print("Мода:", mode_exp)
```

```python
#Завдання 4

# підрахунок респондентів з віддаленою роботою
remote_workers = len(df_public[df_public['RemoteWork'] == 'Fully remote'])

print(remote_workers)
```

```python
#Завдання 5

# Відсоток респондентів, які використовують Python
python_users = df_public['LanguageHaveWorkedWith'].str.contains('Python', na=False).sum()
total_respondents = len(df_public)
percentage = (python_users / total_respondents) * 100

print(f"{percentage:.1f}%")
```

```python
#Завдання 6

# Кількість респондентів, які навчалися через онлайн курси
online_courses_count = df_public['LearnCode'].str.contains('Online Courses', na=False).sum()

print(online_courses_count)
```

```python
#Завдання 7

# Фільтруємо Python-розробників
python_devs = df_public[df_public['LanguageHaveWorkedWith'].str.contains('Python', na=False)]

# Групуємо по країнах і рахуємо статистику
comp_stats = python_devs.groupby('Country')['ConvertedCompYearly'].agg(['mean', 'median'])

print(comp_stats)
```

```python
#Завдання 8

# Сортуємо за компенсацією (спадання) і беремо топ-5
top_5 = df_public.sort_values('ConvertedCompYearly', ascending=False).head(5)

# Виводимо їх освіту
print(top_5[['ConvertedCompYearly', 'EdLevel']])
```

```python
#Завдання 9

# Рахуємо загальну кількість респондентів у кожній віковій категорії
total_by_age = df_public.groupby('Age').size()

# Рахуємо кількість Python-розробників у кожній віковій категорії
python_by_age = df_public[df_public['LanguageHaveWorkedWith'].str.contains('Python', na=False)].groupby('Age').size()

# Рахуємо відсоток
percentage_by_age = (python_by_age / total_by_age * 100).round(1)

print(percentage_by_age)
```

```python
#Завдання 10

# 75-й перцентиль компенсації
percentile_75 = df_public['ConvertedCompYearly'].quantile(0.75)

# Фільтруємо високооплачуваних віддалених працівників
high_paid_remote = df_public[
    (df_public['ConvertedCompYearly'] >= percentile_75) &
    (df_public['RemoteWork'] == 'Remote')
]

# Рахуємо найпопулярніші індустрії
top_industries = high_paid_remote['Industry'].value_counts()

print("Найпопулярніші індустрії серед високооплачуваних віддалених працівників:")
print(top_industries.head(10))  # Показуємо топ-10
```
